## Structural VAR -- Sims example

```python
## import all the modules need to be used
import statsmodels 
import statsmodels.tsa.stattools as tsa
import statsmodels.api as sm
import pandas as pd
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
import matplotlib.pylab
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 8, 3
```

```python
## Load data -- 
df = pd.read_csv('../data/M5_Part II_SVAR_Workfile_Sims.csv')
df['time'] = pd.to_datetime(df.dateid01)
df.index = df['time']
df.drop(['dateid','dateid01'],axis = 1,inplace=True)   ## delete lgdp_gap, we will cancluate it ourselves
```

### steps to conduct SVAR process and get Impulse Response

1. Run the model using VAR reduced form. 
2. Use the all the estimators from reduced form to the Matrix A. 
3. Use matrix a to get the Structural VAR model. 
4. Use SVAR model to do inpulse Response analysis
#### In eviews and statsmodules, process we do not need to do 2-4 manuely. Built package is available to conduct impulse response analysis

### Run Sims example 

##### Firstly, try a simple model, the model is misspecified, we can see the problem when we get the results 

```python
## only use federal funds rate, money supply, price and output 
df.head(2)
```

```python
## first transform variables we want to use 
df['ly']=np.log(df.y)
df['lxr']=np.log(df.xr)
df['lm']=np.log(df.m1)
df['lp']=np.log(df.cpi)
df['lcp']=np.log(df.cmp)
df.head(2)
```

```python
## first, we use only four variables from 1958m4 to 1991m2
data = df[['ff','lm','lp','ly']]['1958-04-01':'1991-02-01'].dropna()

## run VAR model, we are using 14 lags, the same as Sims used
VAR_model = sm.tsa.VAR(data)
res = VAR_model.fit(14)
```

```python
## get the results for VAR reduced form  
res.summary()
```

```python
## get the impulse and response results 
## you can specified how it need to be done, by default 
# VARResults.irf(periods=10, var_decomp=None, var_order=None) --- we can change the order here
## Orthogonalization is done using the Cholesky decomposition of the estimated error covariance matrix 
## \hat \Sigma_u and hence interpretations may change depending on variable ordering.

irf = res.irf(48)
irf.plot(orth=True)


## look at the first chart, we impulse a exorgenous shock to federal funds rate. And the string thing is 
## price goes up in early period. This lead Sims to think the the model is misspecified. Federal funds rate 
## is not completely exorgenous, it could be some leading indicator that caused ff to incrise. we missed 
## other variables here 
## the exorgenous shock is not properly identified
```

```python
irf.plot(impulse='ff') ## only look at the effect of ff to others
irf.plot_cum_effects(orth=False) ## plot the accumulate effects charts 
```

## Now we can try a refined model

```python
## first, we use 6 variables from 1958m4 to 1991m2
data = df[['ff','lxr','lcp','lm','lp','ly']]['1958-04-01':'1991-02-01'].dropna()
## run VAR model, we are using 14 lags, the same as Sims used
VAR_model = sm.tsa.VAR(data)
res = VAR_model.fit(14)
## get the impulse and plot it 
irf = res.irf(48)
irf.plot(orth=True)

## now, we can see the thet price almost stayed the same and falls down in long term
```

### Take a look at Forecast Error Variance Decomposition

```python
fevd = res.fevd(10)
fevd.summary()
```

```python
## we can also visualize forecast error variance
res.fevd(20).plot()
```
