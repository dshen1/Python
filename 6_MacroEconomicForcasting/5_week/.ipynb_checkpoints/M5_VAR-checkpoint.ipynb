# VAR - Reduced Form

```python
## import all the modules need to be used
import statsmodels 
import statsmodels.tsa.stattools as tsa
import statsmodels.api as sm
import pandas as pd
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
import matplotlib.pylab
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 10, 4
```

## Quick demo of how to run VAR - Reduced form

```python
## load sample data  -- Thailand inflation data from 2003m1 to 2014m12
df = pd.read_csv('../data/M5-Part I_VAR_Workfile_CAN.csv')
df['time'] = pd.to_datetime(df.dateid01)
df.index = df['time']
df.drop(['dateid','dateid01','lgdp_gap','lgdp'],axis = 1,inplace=True)   ## delete lgdp_gap, we will cancluate it ourselves
```

```python
### some data transformation, calculate lgdp_gap 
df['lgdp'] = np.log(df.gdp)
cycle,trend = sm.tsa.filters.hpfilter(df.lgdp,1600)
df['lgdp_gap'] = cycle
df['lgdp_trend'] = trend
df.head()
```

```python
## prepare the VAR data 
## there are missing data for infl in the begaining, we need to drop na, eview does the same thing impliciately
data = df[['lgdp_gap','infl','rer_gap','mpr']]['1993-01-01':'2012-10-01'].dropna()

## run VAR model 
VAR_model = sm.tsa.VAR(data)
res = VAR_model.fit(2)
```

```python
## see the summary results, it is the same as eviews output 
res.summary()
```

```python
## take a look the input series 
res.plot()
```

### How to determine what is the correct lag length

1. you can take a look at the auto corrolation for all the independent variables 

```python
## auto corrolation tables
res.plot_acorr()
# looks lik most of them, after 2 or 3 lags, things die down
```

```python
## formally, you can use auto select functions 
## usually if you have monthly data, test 12, quarterly data test 4
VAR_model.select_order(10)

## from this result, i will probably just use 2 lags 
```

## How to test the stationarity of VAR model

```python
## the var result object will return roots, we can plot them 
## to make it the same as Eview, we will take the inverse of it, and see if all of them are inside the unit circly
roots_inve = 1/res.roots
unit_circle = plt.Circle((0, 0), 1, color='r',fill=False)

fig,ax = plt.subplots(figsize=(4,4))
plt.axis([-2, 2, -2, 2])
ax.scatter(roots_inve.real,roots_inve.imag)
ax.add_artist(unit_circle)

## from this result, we can see that roots are all within the circle. so we are good
```

## How we check residuals

```python
## plot the residuals
res.resid.plot(figsize=(9,3))
## in general they seems to be stationary, there are some big variance in financial crisis,they could
## be a proble 
```

There are a lot ways that we can test our residual, see if they are non-serial corrolated, nornal, and homocadastic. 
see: http://statsmodels.sourceforge.net/devel/stats.html

Here we will jsut show how one way, test resid one by one. I can't  mimic Eviews' LM test. need further work on this

```python
## one way is to look a ACF p-values - the test is actually ljungbox test -- or we can just run ljungbox test
test = []
for var_name in res.names:
    #acf = tsa.acf(res.resid[var_name], nlags = 5,qstat=True)
    ljbox = statsmodels.stats.diagnostic.acorr_ljungbox(res.resid[var_name],5)[1]
    p_value = pd.Series(ljbox,name = var_name)
    test.append(p_value)
test_pvalue = pd.DataFrame(test).T
test_pvalue

## here, nothing is significant, so we probably don't have any series corrolation
```

## Forecast using VAR

There is no deterministic/stochastic or dynamic/static solution in statsmodel. I think it is default to be stochastic. and if you want static , use fittedvalues- only work in sample. if you want dynamic forecast, use forecast.

```python
### use the forecast function to do dynamic stochastic forecast
### the way that confidence interval calculate in statsmodels seems to be different from eviews,so it
### won't match up exactly, but the patten shoudl be similar
n_step = 8 
mid, lower, upper = res.forecast_interval(res.y[-res.k_ar:], n_step,alpha = 0.05)
frequency = pd.infer_freq(data.index)
rng = pd.date_range(data.index[-1], periods = n_step+1, freq = frequency)[1:]
df_mid=pd.DataFrame(mid,columns=[x+'_hat' for x in res.names ],index=rng)
df_low=pd.DataFrame(lower,columns=[x+'_low' for x in res.names ],index=rng)
df_up=pd.DataFrame(upper,columns=[x+'_up' for x in res.names ],index=rng)
forecast = pd.concat([df_mid,df_low,df_up],axis =1)
forecast.head()
```

```python
## visualize the predictions #####

result = pd.concat([df[['lgdp_gap','infl','rer_gap','mpr']],forecast],axis=1)
fig = plt.figure(figsize=(14,8))
layout = (2, 2)
lgdp_gap_ax   = plt.subplot2grid(layout, (0, 0))
infl_ax = plt.subplot2grid(layout, (0, 1))
rer_gap_ax  = plt.subplot2grid(layout, (1, 0))
mpr_ax = plt.subplot2grid(layout, (1, 1))
result[['lgdp_gap','lgdp_gap_hat']]['2010':].plot(ax=lgdp_gap_ax)
lgdp_gap_ax.set_title('lgdp_gap')
result[['infl','infl_hat']]['2010':].plot(ax=infl_ax)
infl_ax.set_title('infl')
result[['rer_gap','rer_gap_hat']]['2010':].plot(ax=rer_gap_ax)
rer_gap_ax.set_title('rer_gap')
result[['mpr','mpr_hat']]['2010':].plot(ax=mpr_ax)
mpr_ax.set_title('mpr')
```
