## scrip some online links and download the data

```python
# import modules 

from bs4 import BeautifulSoup
import requests

from pandas import read_html
import pandas as pd
from pandas import Series, DataFrame
from pandas import ExcelWriter

#Lets grab a url for list of failed banks
url = 'http://www.bcrp.gob.pe/statistics/meoc/monetary-and-exchange-operations-daily-tables.html'

result = requests.get(url)
c = result.content

# Set as Beautiful Soup Object 
soup = BeautifulSoup(c)
data = []
idx = []
link = []
name = []


for a in soup.find_all('a',href=True):
    data.append(a['href'])

index = 0 
for item in data:
    if 'xls' in item:
        index += 1
        idx.append(index)
        link.append("http://www.bcrp.gob.pe"+item)
        s = item.split("/")[-1]
        name.append(s)
        
idx = Series(idx)
link = Series(link)
name = Series(name)

print name 
FXI_df = pd.concat([idx,link,name],axis=1)
FXI_df.columns=['Index','Link','Name']

writer = ExcelWriter('output.xlsx')
FXI_df.to_excel(writer,'Sheet1')
writer.save()

print FXI_df
```

```python
# set up the data frame
import urllib
for i,row in FXI_df.iterrows():
    download_url = row['Link']
    file_name = row['Name']
    urllib.urlretrieve(download_url,"./downloads/"+file_name)
```
