```python
#####################################################################
######### Binomial discrete probability distribution ################

# Probility problem 
import scipy.misc as sc                  # use scipy for combination 
import matplotlib.pyplot as plt

p = .72                                  # probability of success
n= 11                                    # number of shots 
k = 6                                    # number of success
comb = sc.comb(n,k)                      # number of n,k combinations 
answer = 100*comb * (p**k)*((1-p)**(n-k))
print 'The probability of make 6 out of 11 shorts is %1.2f%%.' %answer
###### mean and variance ############
from scipy.stats import binom
mean, var = binom.stats(n,p)
print 'mean = %1.2f' %mean 
print 'var = %1.2f' %var 
######## plot the PDF###########
x = range(n+1)                           #indexing starts at 0, so we use n+1 
y = binom.pmf(x,n,p)                     #create the probability mass function
print x 
print y
plt.plot(x,y,'o')                        #plot the PDF 
plt.xlabel('Number of Shoots Made')
plt.ylabel('Probability')
```

```python
#####################################################################
### Poisson Distribution ############################################
import numpy as np
from scipy.stats import poisson

lamb = 10                              # set lambda as the mean of the occurance 
k = np.arange(30)                      # what is the probability the exactly happens k times
pmf_pois = poisson.pmf(k,lamb)         # create the poision distribution values 
plt.bar(k,pmf_pois)                    # plot the poisson distribution
#the probability of 10 or less custom shows up 
prob_10 = poisson.cdf(10,lamb)
prob_O10 = 1 - prob_10
print 'The probability that 10 or less customers show up is %0.1f %%' %(100*prob_10)
print 'The probability that more than 10 customers show up is %0.1f %%' %(100*prob_O10)
```

```python
######################################################################
############# Normal distribution ####################################
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline

x = np.arange(-5,5,0.01)                      # create a range from -5 to 5, step 0.01  --- this is just the x axis
                                              # not the real distribution 
mean = 0                                      # set the mean to be 0 
std = 1                                       # set standard deviation to be 1
y = stats.norm.pdf(x,mean,std)                # create the distribution using normal distribution function
plt.plot(x,y)                                 # plot the normal distribution 
```

```python
#### create a distribution by get random 1000 number from a normal distribution #########
sigma = 1                                     # set the mean and standard deviation
mu = 0                                
norm_set = np.random.normal(mu,sigma,1000)    # generating a list with random 1000 values from a normal distribution 
plt.hist(norm_set,bins=50)
```

```python
#####################################################
######## T distribution #############################
from scipy.stats import t 

x = np.linspace(-5,5,100)                           # linear spaced, from -5 to 5, for 100 values
rv = t(5)                                           # create the t distribution with sample size 3
plt.plot(x,rv.pdf(x))                              # plot the t distribution 
```
