```python
##################################
## Machine Learning Intro #######
#################################

######### linear regression #########

import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
from sklearn.datasets import load_boston                       #import sample data sets from sklearn 
boston=load_boston()
print boston.DESCR
```

```python
plt.hist(boston.target,bins=50)                  #take a look as the distribution of the medium house price
plt.xlabel('Prices in $1000s')
plt.ylabel('Number of houses')
```

```python
#make boston data sets into a data frame 
df = DataFrame(boston.data)                        #transform array into a data frame
df.columns = boston.feature_names                 #rename colums index name
df['Price'] = boston.target
df.head()
```

```python
plt.scatter(df.RM,boston.target)      #plot number of rooms against house price 
plt.ylabel('Price in $1000s')
plt.xlabel('Number of rooms')
```

```python
sns.lmplot('RM','Price',data=df)       # plot with linear fit 
```

```python
#####################################################################
######## Using Numpy for Univariate Linear Regression ###############

# numpy need to use a  two-dimensional arrary to perform linear algebra 
X = np.vstack(df.RM)                          # transform datafram to a numpy vector or matrix 
X = np.array([[value,1] for value in X])      # make a [X,1] matrix for performing regression in numpy later
Y = df.Price

m,b = np.linalg.lstsq(X,Y)[0]                 # get the coefficient m and constant b 
print 'coefficient is %.2f' %m
print 'constant is %.2f' %b

#draw the scatter plot and best fit line manually
plt.plot(df.RM,df.Price,'o')                  # scatter plot 
x=df.RM
plt.plot(x,m*x+b,'r',label='BEst FIt Line')   # fited line 
```

```python
############ get the residual of the regression ##########

result = np.linalg.lstsq(X,Y)           # get the output of the regression
error_total = result[1]                 # get the total error
rmse = np.sqrt(error_total/len(X))      # get the root mean square
print "The root mean squared error was %.2f" %rmse
```

```python
####################################################
#### use sklearn to perfrom linear regressions #####
####################################################
import sklearn
from sklearn.linear_model import LinearRegression 

lreg = LinearRegression()
X_multi = df.drop('Price',1)                         # drop y, Price variable 
Y_target = df.Price                                  # set y to be Price
lreg.fit(X_multi,Y_target)                           # run the regression

### print out the result 
print 'the intercept is %.2f' %lreg.intercept_
coeff = DataFrame(df.columns)
coeff.columns=['Variables']
coeff['Coefficients'] = pd.Series(lreg.coef_)
print coeff                                          # print out the coefficients table
```

```python
#######################################################
######## Using Training and Validation ################
#######################################################

X = np.vstack(df.RM)                                                            #transform x to be a vector 
Y = np.vstack(df.Price)                                                         #transform x to be a vector
X_train, X_test,Y_train,Y_test = sklearn.cross_validation.train_test_split(X,Y) #X is RM vector
lreg= LinearRegression()
lreg.fit(X_train,Y_train)
pred_train = lreg.predict(X_train)
pred_test = lreg.predict(X_test)
print "Fit a model X_ train , and calculate the MSE with Y _train: %.2f " %np.mean((Y_train-pred_train)**2)
print "Fit a model X_train, and calculate MSE with X_test and Y_test: %.2f" %np.mean((Y_test-pred_test)**2)

### residule plot ###
train = plt.scatter(pred_train,(pred_train -Y_train),c='b',alpha=0.5)
test = plt.scatter(pred_test,(pred_test-Y_test),c='r',alpha=0.5)
plt.hlines(y=0,xmin=-10,xmax=40)
plt.legend((train,test),('Training','Test'),loc='lower left')
plt.title('Residual Plots')
```
