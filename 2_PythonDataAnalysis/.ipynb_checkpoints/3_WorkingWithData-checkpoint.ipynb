```python
##########
## merge
##########
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

df1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
df2 = DataFrame({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})

## many-to-one merge 
pd.merge(df1,df2,on='key')                   #only merged togeterh where keys are in common 
pd.merge(df1,df2,on='key',how='left')              #only keep left
pd.merge(df1,df2,on='key',how='right')             #only keep right 
pd.merge(df1,df2,on='key',how='outer')             #the union of both keys, merge and add addition keys

### merge on multiple columns ####
df3 = DataFrame({'key1': ['X', 'y', 'z', 'a', 'b', 'c'],
                 'key2':[1,2,3,4,5,6],
                 'data_set_3': range(6)})
df4 = DataFrame({'key1': ['Y', 'a', 'X', 'b', 'Z'],
                 'key2':[2,4,1,3,5],
                 'data_set_4': range(5)})

pd.merge(df3,df4,on=['key1','key2'],how='left')                          #keep only left, on multiple keys
pd.merge(df3,df4,on=['key1'],how='left')                                 #pandans will add suffixex to the same column 

# For more info on merge parameters check out:
# url = 'http://pandas.pydata.org/pandas-docs/dev/generated/pandas.DataFrame.merge.html'

################## merge on indexes ###########
df_group = DataFrame({'group_data': [10, 20]}, index=['X', 'Y'])
pd.merge(df1,df_group,left_on='key',right_index=True,how='outer')         #merge using keys and index

###use join 
df1.join(df_group)                                                        #can use join as well 
```

```python
##########
## Duplicates 
##########
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

df= DataFrame({'key1':['A']*2 + ['B']*3,
              'key2':[2,2,2,3,3]})
df.duplicated()                           #duplicates of all variables 
df.drop_duplicates()                      #drop duplicates 
df.drop_duplicates('key1')                #drop dup on key1
df.drop_duplicates('key1',take_last=True) #drop dup ok key1 take the last value
```

```python
########
# mapping  and replace 
########
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

###mapping ##
df = DataFrame({'city':['Alma','Brian Head','Fox Park'],
                    'altitude':[3158,3000,2762]})
state_map={'Alma':'Colorado','Brian Head':'Utah','Fox Park':'Wyoming'}
df['state'] = df['city'].map(state_map)             #add state and map to citiets 

##replace##
df.replace('Alma','replaced')                       #replace a string
df.altitude.replace([3158,3000],[111,222],inplace=True)          #replace multiple values 

### rename index ####
df.columns = df.columns.map(str.upper)              #make all column index to upper case
df.rename(columns={'CITY':'City'},inplace = True)   #change columns name 

#### bins #########
years = [1990,1991,1992,2008,2012,2015,1987,1969,2013,2008,1999]
decade_bins = [1960,1970,1980,1990,2000,2010,2020]
decade_cat = pd.cut(years,decade_bins)
decade_cat.categories
pd.value_counts(decade_cat)             #count element in each bin
pd.cut(years,2,precision=1)             #make data into two bins , precision is 1 year 

#### outlires #######
np.random.seed(12345)
df=DataFrame(np.random.randn(1000,4))    #made a df with random number into 4 columns 
df[(np.abs(df)>3).any(1)]                #any variable has abs > 3, return the row 
df[np.abs(df)>3] = np.sign(df)*3         #assign value to selected value, make outliers to 3 or -3 
df.describe()

####### Permutation ########
df=DataFrame(np.arange(16).reshape(4,4))
blender = np.random.permutation(4)                  # a list of random order from o to 4 
df.take(blender)                                    #reorder the df based on blender order 
### simulate a random draw 
box = np.array([1,2,3])
shaker = np.random.randint(0,len(box),size=10)      #create a random order ofr 0 to 2 list for random draw
hand_grabs = box.take(shaker)                       #randow draw 10 times from a box 
print hand_grabs
```

```python
##############################################
############### group by #####################
##############################################
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

df = pd.read_csv('data.csv',sep=',',                              
                 usecols=['country','year','PPPPC',               
                          'rule_law','control_corruption',
                          'regulatory_quality'],    #only read certain columns into the dataframe         
                nrows=10                            #only read certain rows
                ) 
#print df.head()
df['k1'] = df['PPPPC']>7000                         #generate an indicator variable for group
df['k2'] = df['rule_law'] > 0                       #generate second indicator
group1 = df['PPPPC'].groupby(df.k1)                 #make group object based on indicator
group2 = df['PPPPC'].groupby([df.k1,df.k2])         #make group object based on two indicators 
group2.mean()                                       #mean by groups 

#### df group by 
df.groupby('k1').mean()                             #all bariables based on indicators 
df.groupby(['k1','k2']).mean()                      #based on multiple indicators 
df.groupby('k1').size()                             #the count in the groups 

##### iterate with groups
for (k1,k2), group in df.groupby(['k1','k2']):
    print "key1 = %r, key2 = %r" %(k1,k2)
    print group.head() 
    print '\n'
    
## make group into dictionary
group_dict = dict(list(df.groupby('k1')))          #make dict with groups
group_dict[True]                                   #only show the df with key = True
```

```python
#######################
### Data aggregation
#######################
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
%matplotlib inline

########## aggregate data #########
df = pd.read_csv('data.csv',sep=',',                              
                 usecols=['country','year','PPPPC',               
                          'rule_law','control_corruption',
                          'regulatory_quality','state_reform'],    #only read certain columns into the dataframe         
                nrows=100                                          #only read certain rows
                ) 

group_df = df.groupby('state_reform')
group_df.describe()                                             #describe by group 
group_df.agg('mean')                                            #only get the mean

### customized aggregations
def max_to_min(arr):
    return arr.max()-arr.min()
group_df.agg(max_to_min)                                        #use customized aggregation method
df['all_index'] = df[['rule_law','control_corruption',          #created an index using row total
                      'regulatory_quality',
                      'state_reform']].sum(axis=1)
df.pivot_table(index=['all_index'])                             #the same as gorup by 
df.plot(kind='scatter',x='PPPPC',y='all_index')                 #scatter plot 

#### split apply combine #########
df['HI'] = np.where(df['PPPPC']>df['PPPPC'].mean(),1,0)          #use np.where to assing indicators
df.loc[df.PPPPC>df.PPPPC.quantile(q=0.75),'HI'] = 2              #use loc to assing part of a df to certain value, based on certain value
    #create a ranker function, for latter apply, similary to by: egen function in stata
def ranker(df):
    df['all_index_rank'] = np.arange(len(df)) +1 
    return df
df.sort_values('all_index', ascending = False, inplace = True)   #sort desending 
df = df.groupby('HI').apply(ranker)                              #rank based on all_index, group by HI
                                                                 #similar to stata by: egen etc
df.sort_values(['HI','all_index'])

##### cross tabulation ####
pd.crosstab(df.state_reform, df.HI,margins=True)                 #similar to stata tab
```
