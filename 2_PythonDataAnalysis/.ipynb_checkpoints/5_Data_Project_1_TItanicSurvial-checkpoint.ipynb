```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
###### see how is on board #############
sns.factorplot(x='Pclass',data=df,kind='count',hue='Sex')            #passenger by gender and class
```

```python
def male_female_child(passenger):
    age,sex = passenger
    if age<16:
        return 'child'
    else:
        return sex
```

```python
#use apply to generate new columns
df['person']=df[['Age','Sex']].apply(male_female_child,axis=1)             #generage a new column with male, female and child
print df.head(10)['person']
sns.factorplot(x='Pclass',data=df,kind='count',hue='person') 
```

```python
##show histogram of age 
df.Age.hist(bins=50)                                                     #see age distribution
```

```python
df.Age.mean()                                                           #see means
```

```python
df.person.value_counts()                                                #count by group
```

```python
##take a look at the kde plot , see how age distributed by certain group

fig =sns.FacetGrid(data=df,hue='Sex',aspect=4)            #make the fig aspect ratio to be 4
fig.map(sns.kdeplot,'Age',shade=True)
oldest = df.Age.max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
## age distribution by class

fig =sns.FacetGrid(data=df,hue='Pclass',aspect=4)            
fig.map(sns.kdeplot,'Age',shade=True)
oldest = df.Age.max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
# age distribution by sex and child
fig =sns.FacetGrid(data=df,hue='person',aspect=4)            
fig.map(sns.kdeplot,'Age',shade=True)
oldest = df.Age.max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
### look at deck,see how cambins distributed by levels
deck = df.Cabin.dropna()                         #drop all missings
#create a list to get the first letter of Cabin
cabin_df = DataFrame(deck)
cabin_df['levels']=cabin_df.Cabin.str[:1]
cabin_df=cabin_df[cabin_df.levels!='T']                                                    #drop rows if levels = T
cabin_df.sort_values(by='levels',inplace=True)                                             #sort by column levels
print cabin_df.head()
sns.factorplot('levels',data=cabin_df,palette='winter',kind='count')                     #number of cambins by level
```

```python
############## who is alone and who is with family 
df.head()
df['alone'] = df['SibSp'] + df['Parch']
df['alone'].loc[df['alone']>0]='With Family'
df['alone'].loc[df['alone']==0] = 'Alone'
sns.factorplot('alone',data=df,palette='Blues',kind='count')
```

```python
########look at survivers #############
df['Survivor']=df.Survived.map({0:'no',1:'yes'})
sns.factorplot('Survivor',data=df,palette='Set1',kind='count')
```

```python
###### survior by group #############
sns.factorplot('Survivor',data=df,hue='Pclass',palette='Set1',kind='count')
```

```python
sns.factorplot(x='Pclass',y='Survived',hue='person',data=df)
```
